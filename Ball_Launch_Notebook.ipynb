## 1. Create a Model of ball launched  in the horizontal direction 
### Create a computational model of ball launched with some velocity in the horizontal direction and create of the 2D trajectory of the ball.
Your model should include the mass and size of the ball, the initial horizontal speed, and the initial height of the ball above the floor. The model should stop when the ball hits the floor. The model should return the horizontal distance the ball travels before hitting the floor as well as the total time it is in the air.  

You will need to edit the code at the point indicated below.   
Pseudo code:  
$x_{i+1}=x_{i}+v_{x,i} \Delta t$  
$y_{i+1}=y_{i}+v_{y,i} \Delta t$  
$v_{y,i+1}=v_{y,i}-g \Delta t$  


```python
#Sarah Phan-Budd
#Physics Department, Winona State University
#sphanbudd@winona.edu

#Modified from:
#Title: Projectile Motion: Experiment and Computational Model(PICUP Code Library)
#Author: Todd Zimmerman
#Date: 2017
#Code version: Published in the PICUP Collection, July 2016
#Availiblity: https://www.compadre.org/PICUP/exercises/Exercise.cfm?A=proj_mot&S=3

#Thanks to Carl Ferkinhoff, Winona State University, for lab template
from math import *
import matplotlib.pyplot as plt
%matplotlib inline

dt = 0.01    #dt is the time step 'Delta t'
t=0.   #Set initial time to zero
g = [0.,-9.8,0.]
h = 2.   #initial height of ball
v_init=12.
ball_pos=[0.,h,0.] #ball_pos[0] is x, ball_pos[1] is y and ball_pos[2] is z
ball_v=[v_init,0.,0.]

x_array=[ball_pos[0]]
y_array=[ball_pos[1]]
while ball_pos[1]>0:  #Stop when ball hits floor
    for i in range(0, 3):
        #edit the following lines of code
        ball_pos[i] = ball_pos[i]-10*dt  #Update position of ball #this is not the right equation
        ball_v[i]=10 #this is also the right equation
        #this is the end of the portion of the code you will need to edit 
        
    x_array.append(ball_pos[0]) 
    y_array.append(ball_pos[1])  
    #print(ball_pos[1])
    t+=dt   #Calculate total time elapsed

print("Time elapsed = %1f" %t)
print("Horizontal distance traveled = %1f" %ball_pos[0])
plt.plot(x_array,y_array)
plt.xlabel('Vertical Position (m)')
plt.ylabel('Horizontal Position (m)')
plt.xlim((0.0, ball_pos[0]))
```
