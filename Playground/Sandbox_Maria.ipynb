```python
import matplotlib.pyplot as plt # get matplotlib to plot 
```

```python
t, h, n = 0.0, 0.1, 20          # init time, step size, number of ticks
```

```python
v0 = float(input("enter initial velocity:"))
```

```python
ta, xa = [], []                 # time and position arrays for plotting
```

```python
for i in range(n):              # loop for n steps
    ta.append(t)                # record time
    xa.append(v0*t)             # update position
    t = t + h                   # update time 
```

```python
plt.figure()                    # start a figure;
plt.plot(ta, xa, '-o')          # plot data, '-o' means blue
plt.xlabel('t (s)')             # add labels
plt.ylabel('x (m)')
plt.show()                      # show figure
```
