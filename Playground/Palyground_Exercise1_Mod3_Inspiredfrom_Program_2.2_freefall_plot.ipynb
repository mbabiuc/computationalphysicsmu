```python
#
# Program 2.2: Free fall with Euler's method (freefall_plot.ipynb)
# J Wang, Computational modeling and visualization with Python
#

import matplotlib.pyplot as plt # get matplotlib plot functions
%matplotlib notebook
```

```python
def motion(v0, a, dt, tf):
    v = v0
    t, y = 0.0, 0.0                 # initial time and position (multiple assignments on the same line)
    tnum, ynum, vnum = [], [], []   # time and numerical position arrays for plotting
    yana = []                       # array holding analytic solution

    while t < tf:        # loop for one second
# Euler numerical method
        v = v0 - a*t                
        y = y + v*dt 
# End Euler Method        

        vnum.append(v)
        ynum.append(y)
        yana.append(v0*t-a*t*t/2.0)   # analytic solution for position

        t = t + dt
        tnum.append(t)    # record time and position
        
    plt.figure()                    # start a figure
    plt.plot(tnum,ynum, tnum,yana,'--')     # draw 2nd curve as dashed
    plt.xlabel('t (s)')             # add labels
    plt.ylabel('y (m)')
    plt.show()                      # show figure
```

```python
motion(v0=5, a=9.8, dt=0.1, tf=1)
```

```python
motion(v0=5, a=9.8, dt=0.01,tf=1)
```

```python
g = 0.0
dt = 0.01                       # the time step size
v0 = 5.0
t, y = 0.0, 0.0                 # initial time and position (multiple assignments on the same line)
tnum, ynum, vnum = [], [], []   # time and numerical position arrays for plotting
yana = []                       # array holding analytic solution
```

```python
v = v0
while t<1.0:        # loop for one second
    t = t + dt
    y = y + v*dt                  # Euler's method
    v = v0 - g*dt                 # calc v(t)
    tnum.append(t)    # record time and position
    ynum.append(y)
    yana.append(v0*t-g*t*t/2.0)   # analytic solution for position
```

```python
plt.figure()                    # start a figure
plt.plot(tnum,ynum, tnum,yana,'--')     # draw 2nd curve as dashed
plt.xlabel('t (s)')             # add labels
plt.ylabel('y (m)')
plt.show()                      # show figure
```
